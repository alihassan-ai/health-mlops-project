```python
"""
Exploratory ../data Analysis - Health MLOps Project
Analyzes health, air quality, and weather ../data to find patterns
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
import os

warnings.filterwarnings('ignore')

# Set style for better looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 6)

print("=" * 80)
print("EXPLORATORY ../data ANALYSIS - HEALTH MLOPS PROJECT")
print("=" * 80)

# Create output directory for plots
os.makedirs('notebooks/eda_plots', exist_ok=True)

# ============================================================================
# PART 1: LOAD ALL ../dataSETS
# ============================================================================
print("\n[PART 1] Loading ../datasets...")
print("-" * 80)

# Load Health ../data from all nodes
health_dfs = []
for node in range(1, 6):
    df = pd.read_csv(f'../../../../data/raw/health/node_{node}_health_../data.csv')
    health_dfs.append(df)
    print(f"✓ Loaded node_{node}_health_../data.csv: {len(df):,} records")

df_health = pd.concat(health_dfs, ignore_index=True)
print(f"\n✓ Total Health Records: {len(df_health):,}")

# Load Air Quality ../data
cities = ['New_York', 'Los_Angeles', 'Chicago', 'Houston', 'Phoenix']
air_dfs = []
for city in cities:
    df = pd.read_csv(f'../../../data/raw/air_quality/{city}_air_quality.csv')
    air_dfs.append(df)
    print(f"✓ Loaded {city}_air_quality.csv: {len(df):,} records")

df_air = pd.concat(air_dfs, ignore_index=True)
print(f"\n✓ Total Air Quality Records: {len(df_air):,}")

# Load Weather ../data
weather_dfs = []
for city in cities:
    df = pd.read_csv(f'../../../data/raw/weather/{city}_weather.csv')
    weather_dfs.append(df)
    print(f"✓ Loaded {city}_weather.csv: {len(df):,} records")

df_weather = pd.concat(weather_dfs, ignore_index=True)
print(f"\n✓ Total Weather Records: {len(df_weather):,}")

# ============================================================================
# PART 2: BASIC ../data EXPLORATION
# ============================================================================
print("\n" + "=" * 80)
print("[PART 2] Basic ../data Exploration")
print("=" * 80)

print("\n--- HEALTH ../data ---")
print(df_health.info())
print("\nFirst few rows:")
print(df_health.head())
print("\nBasic Statistics:")
print(df_health.describe())
print("\nMissing Values:")
print(df_health.isnull().sum())

print("\n--- AIR QUALITY ../data ---")
print(df_air.info())
print("\nFirst few rows:")
print(df_air.head())
print("\nBasic Statistics:")
print(df_air.describe())

print("\n--- WEATHER ../data ---")
print(df_weather.info())
print("\nFirst few rows:")
print(df_weather.head())
print("\nBasic Statistics:")
print(df_weather.describe())

# ============================================================================
# PART 3: HEALTH ../data ANALYSIS
# ============================================================================
print("\n" + "=" * 80)
print("[PART 3] Health ../data Analysis")
print("=" * 80)

# Health Status Distribution
print("\n1. Health Status Distribution:")
print(df_health['health_status'].value_counts())
print("\nPercentages:")
print(df_health['health_status'].value_counts(normalize=True) * 100)

# Plot Health Status Distribution
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Bar plot
df_health['health_status'].value_counts().plot(kind='bar', ax=axes[0], color=['green', 'orange', 'red'])
axes[0].set_title('Health Status Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Health Status')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Pie chart
df_health['health_status'].value_counts().plot(kind='pie', ax=axes[1], autopct='%1.1f%%', colors=['green', 'orange', 'red'])
axes[1].set_title('Health Status Proportion', fontsize=14, fontweight='bold')
axes[1].set_ylabel('')

plt.tight_layout()
plt.savefig('notebooks/eda_plots/01_health_status_distribution.png', dpi=300, bbox_inches='tight')
print("\n✓ Saved: 01_health_status_distribution.png")
plt.show()

# Health Metrics by Status
print("\n2. Health Metrics by Status:")
health_metrics_by_status = df_health.groupby('health_status')[['heart_rate', 'spo2', 'body_temp']].mean()
print(health_metrics_by_status)

# Visualize health metrics by status
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, metric in enumerate(['heart_rate', 'spo2', 'body_temp']):
    df_health.boxplot(column=metric, by='health_status', ax=axes[idx])
    axes[idx].set_title(f'{metric.replace("_", " ").title()} by Health Status', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('Health Status')
    axes[idx].set_ylabel(metric.replace("_", " ").title())
    plt.sca(axes[idx])
    plt.xticks(rotation=45)

plt.suptitle('')
plt.tight_layout()
plt.savefig('notebooks/eda_plots/02_health_metrics_by_status.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 02_health_metrics_by_status.png")
plt.show()

# Distribution of vital signs
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
metrics = ['heart_rate', 'steps', 'sleep_hours', 'spo2', 'body_temp']

for idx, metric in enumerate(metrics):
    row = idx // 3
    col = idx % 3
    
    # Histogram
    axes[row, col].hist(df_health[metric], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
    axes[row, col].set_title(f'Distribution of {metric.replace("_", " ").title()}', fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel(metric.replace("_", " ").title())
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].grid(alpha=0.3)

# Remove empty subplot
fig.delaxes(axes[1, 2])

plt.tight_layout()
plt.savefig('notebooks/eda_plots/03_vital_signs_distribution.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 03_vital_signs_distribution.png")
plt.show()

# Correlation between health metrics
print("\n3. Correlation between Health Metrics:")
health_corr = df_health[['heart_rate', 'steps', 'sleep_hours', 'spo2', 'body_temp']].corr()
print(health_corr)

plt.figure(figsize=(10, 8))
sns.heatmap(health_corr, annot=True, cmap='coolwarm', center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix - Health Metrics', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('notebooks/eda_plots/04_health_correlation_matrix.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 04_health_correlation_matrix.png")
plt.show()

# ============================================================================
# PART 4: AIR QUALITY ANALYSIS
# ============================================================================
print("\n" + "=" * 80)
print("[PART 4] Air Quality Analysis")
print("=" * 80)

# Convert timestamp to datetime
df_air['timestamp'] = pd.to_datetime(df_air['timestamp'])
df_air['date'] = df_air['timestamp'].dt.date
df_air['hour'] = df_air['timestamp'].dt.hour

print("\n1. Air Quality Statistics by City:")
air_by_city = df_air.groupby('city_id')[['pm25', 'pm10', 'aqi']].mean()
print(air_by_city)

# Plot air quality by city
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, metric in enumerate(['pm25', 'pm10', 'aqi']):
    air_by_city[metric].plot(kind='bar', ax=axes[idx], color='coral')
    axes[idx].set_title(f'Average {metric.upper()} by City', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('City')
    axes[idx].set_ylabel(f'{metric.upper()}')
    axes[idx].tick_params(axis='x', rotation=45)
    axes[idx].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('notebooks/eda_plots/05_air_quality_by_city.png', dpi=300, bbox_inches='tight')
print("\n✓ Saved: 05_air_quality_by_city.png")
plt.show()

# Hourly pollution patterns
print("\n2. Hourly Pollution Patterns (Rush Hour Effect):")
hourly_pollution = df_air.groupby('hour')[['pm25', 'aqi']].mean()
print(hourly_pollution)

plt.figure(figsize=(14, 6))
plt.plot(hourly_pollution.index, hourly_pollution['pm25'], marker='o', linewidth=2, label='PM2.5')
plt.plot(hourly_pollution.index, hourly_pollution['aqi'], marker='s', linewidth=2, label='AQI')
plt.axvspan(7, 9, alpha=0.2, color='red', label='Morning Rush')
plt.axvspan(17, 19, alpha=0.2, color='orange', label='Evening Rush')
plt.title('Average Pollution Levels by Hour of Day', fontsize=14, fontweight='bold')
plt.xlabel('Hour of Day')
plt.ylabel('Pollution Level')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('notebooks/eda_plots/06_hourly_pollution_pattern.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 06_hourly_pollution_pattern.png")
plt.show()

# Air quality distribution
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
air_metrics = ['pm25', 'pm10', 'no2', 'co', 'o3', 'aqi']

for idx, metric in enumerate(air_metrics):
    row = idx // 3
    col = idx % 3
    
    axes[row, col].hist(df_air[metric], bins=50, color='lightcoral', edgecolor='black', alpha=0.7)
    axes[row, col].set_title(f'Distribution of {metric.upper()}', fontsize=12, fontweight='bold')
    axes[row, col].set_xlabel(metric.upper())
    axes[row, col].set_ylabel('Frequency')
    axes[row, col].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('notebooks/eda_plots/07_air_quality_distribution.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 07_air_quality_distribution.png")
plt.show()

# ============================================================================
# PART 5: WEATHER ANALYSIS
# ============================================================================
print("\n" + "=" * 80)
print("[PART 5] Weather Analysis")
print("=" * 80)

# Convert timestamp to datetime
df_weather['timestamp'] = pd.to_datetime(df_weather['timestamp'])
df_weather['date'] = df_weather['timestamp'].dt.date
df_weather['hour'] = df_weather['timestamp'].dt.hour

print("\n1. Weather Statistics by City:")
weather_by_city = df_weather.groupby('city_id')[['temperature', 'humidity', 'precipitation']].mean()
print(weather_by_city)

# Plot weather by city
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for idx, metric in enumerate(['temperature', 'humidity', 'precipitation']):
    weather_by_city[metric].plot(kind='bar', ax=axes[idx], color='steelblue')
    axes[idx].set_title(f'Average {metric.title()} by City', fontsize=12, fontweight='bold')
    axes[idx].set_xlabel('City')
    axes[idx].set_ylabel(metric.title())
    axes[idx].tick_params(axis='x', rotation=45)
    axes[idx].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('notebooks/eda_plots/08_weather_by_city.png', dpi=300, bbox_inches='tight')
print("\n✓ Saved: 08_weather_by_city.png")
plt.show()

# Temperature pattern by hour
hourly_temp = df_weather.groupby('hour')['temperature'].mean()

plt.figure(figsize=(14, 6))
plt.plot(hourly_temp.index, hourly_temp.values, marker='o', linewidth=2, color='darkorange')
plt.title('Average Temperature by Hour of Day', fontsize=14, fontweight='bold')
plt.xlabel('Hour of Day')
plt.ylabel('Temperature (°C)')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('notebooks/eda_plots/09_hourly_temperature.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 09_hourly_temperature.png")
plt.show()

# Weather correlation
weather_corr = df_weather[['temperature', 'humidity', 'pressure', 'wind_speed', 'precipitation']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(weather_corr, annot=True, cmap='viridis', center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix - Weather Metrics', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('notebooks/eda_plots/10_weather_correlation_matrix.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 10_weather_correlation_matrix.png")
plt.show()

# ============================================================================
# PART 6: CROSS-../dataSET ANALYSIS
# ============================================================================
print("\n" + "=" * 80)
print("[PART 6] Cross-../dataset Analysis - Finding Relationships")
print("=" * 80)

# Prepare ../data for merging - aggregate by date and city
df_health['date'] = pd.to_datetime(df_health['date'])

# Map hospital nodes to cities
node_to_city = {
    'hospital_1': 'city_1',
    'hospital_2': 'city_2',
    'hospital_3': 'city_3',
    'hospital_4': 'city_4',
    'hospital_5': 'city_5'
}
df_health['city_id'] = df_health['node_id'].map(node_to_city)

# Aggregate health ../data by date and city
health_daily = df_health.groupby(['date', 'city_id']).agg({
    'heart_rate': 'mean',
    'spo2': 'mean',
    'body_temp': 'mean',
    'health_status': lambda x: (x == 'sick').sum() / len(x) * 100  # % sick
}).reset_index()
health_daily.rename(columns={'health_status': 'sick_percentage'}, inplace=True)

# Aggregate air quality by date and city
air_daily = df_air.groupby(['date', 'city_id']).agg({
    'pm25': 'mean',
    'pm10': 'mean',
    'aqi': 'mean'
}).reset_index()
air_daily['date'] = pd.to_datetime(air_daily['date'])

# Aggregate weather by date and city
weather_daily = df_weather.groupby(['date', 'city_id']).agg({
    'temperature': 'mean',
    'humidity': 'mean',
    'precipitation': 'sum'
}).reset_index()
weather_daily['date'] = pd.to_datetime(weather_daily['date'])

# Merge all ../datasets
merged = health_daily.merge(air_daily, on=['date', 'city_id'], how='inner')
merged = merged.merge(weather_daily, on=['date', 'city_id'], how='inner')

print(f"\n✓ Merged ../dataset created: {len(merged)} records")
print("\nMerged ../data Preview:")
print(merged.head())

# Save merged ../data
merged.to_csv('../../data/processed/merged_daily_../data.csv', index=False)
print("\n✓ Saved merged ../data: ../data/processed/merged_daily_../data.csv")

# Correlation analysis across all features
print("\n1. Correlation: Air Quality vs Health")
health_air_corr = merged[['pm25', 'aqi', 'sick_percentage', 'heart_rate', 'spo2', 'body_temp']].corr()
print(health_air_corr)

plt.figure(figsize=(10, 8))
sns.heatmap(health_air_corr, annot=True, cmap='RdYlGn_r', center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation: Air Quality vs Health Metrics', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('notebooks/eda_plots/11_air_health_correlation.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 11_air_health_correlation.png")
plt.show()

# Scatter plots: pollution vs health
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].scatter(merged['pm25'], merged['sick_percentage'], alpha=0.5, color='red')
axes[0].set_xlabel('PM2.5 Level')
axes[0].set_ylabel('% Sick Population')
axes[0].set_title('PM2.5 vs Sick Population %', fontweight='bold')
axes[0].grid(alpha=0.3)

axes[1].scatter(merged['aqi'], merged['heart_rate'], alpha=0.5, color='orange')
axes[1].set_xlabel('AQI')
axes[1].set_ylabel('Average Heart Rate')
axes[1].set_title('AQI vs Heart Rate', fontweight='bold')
axes[1].grid(alpha=0.3)

axes[2].scatter(merged['temperature'], merged['sick_percentage'], alpha=0.5, color='blue')
axes[2].set_xlabel('Temperature (°C)')
axes[2].set_ylabel('% Sick Population')
axes[2].set_title('Temperature vs Sick Population %', fontweight='bold')
axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('notebooks/eda_plots/12_environment_health_scatter.png', dpi=300, bbox_inches='tight')
print("✓ Saved: 12_environment_health_scatter.png")
plt.show()

# ============================================================================
# SUMMARY & KEY INSIGHTS
# ============================================================================
print("\n" + "=" * 80)
print("EDA SUMMARY & KEY INSIGHTS")
print("=" * 80)

print("\n📊 ../dataSET STATISTICS:")
print(f"  • Health Records: {len(df_health):,}")
print(f"  • Air Quality Records: {len(df_air):,}")
print(f"  • Weather Records: {len(df_weather):,}")
print(f"  • Merged Daily Records: {len(merged):,}")

print("\n🏥 HEALTH INSIGHTS:")
print(f"  • Healthy: {(df_health['health_status'] == 'healthy').sum() / len(df_health) * 100:.1f}%")
print(f"  • At Risk: {(df_health['health_status'] == 'at_risk').sum() / len(df_health) * 100:.1f}%")
print(f"  • Sick: {(df_health['health_status'] == 'sick').sum() / len(df_health) * 100:.1f}%")
print(f"  • Avg Heart Rate (Sick): {df_health[df_health['health_status']=='sick']['heart_rate'].mean():.1f} bpm")
print(f"  • Avg Heart Rate (Healthy): {df_health[df_health['health_status']=='healthy']['heart_rate'].mean():.1f} bpm")

print("\n🌫️ AIR QUALITY INSIGHTS:")
print(f"  • Average PM2.5: {df_air['pm25'].mean():.2f} µg/m³")
print(f"  • Average AQI: {df_air['aqi'].mean():.2f}")
print(f"  • Peak pollution hours: 7-9 AM, 5-7 PM (rush hours)")

print("\n🌤️ WEATHER INSIGHTS:")
print(f"  • Avg Temperature Range: {df_weather['temperature'].min():.1f}°C - {df_weather['temperature'].max():.1f}°C")
print(f"  • Avg Humidity: {df_weather['humidity'].mean():.1f}%")

print("\n🔗 KEY CORRELATIONS:")
sick_pm25_corr = merged[['pm25', 'sick_percentage']].corr().iloc[0, 1]
sick_temp_corr = merged[['temperature', 'sick_percentage']].corr().iloc[0, 1]
print(f"  • PM2.5 ↔ Sick %: {sick_pm25_corr:.3f}")
print(f"  • Temperature ↔ Sick %: {sick_temp_corr:.3f}")
print(f"  • AQI ↔ Heart Rate: {merged[['aqi', 'heart_rate']].corr().iloc[0, 1]:.3f}")

print("\n✅ All plots saved in: notebooks/eda_plots/")
print("\n" + "=" * 80)
print("EDA COMPLETE! Ready for Feature Engineering & Modeling")
print("=" * 80)
```

```python
"""
Fixed Data Merging Script - Health MLOps Project
Properly merges health, air quality, and weather data
"""

import pandas as pd
import numpy as np
from datetime import datetime

print("=" * 80)
print("FIXING DATA MERGE - Health MLOps Project")
print("=" * 80)

# ============================================================================
# STEP 1: Load all datasets
# ============================================================================
print("\n[STEP 1] Loading datasets...")

# Load Health Data
health_dfs = []
for node in range(1, 6):
    df = pd.read_csv(f'../data/raw/health/node_{node}_health_data.csv')
    health_dfs.append(df)
    print(f"✓ Loaded node_{node}: {len(df):,} records")

df_health = pd.concat(health_dfs, ignore_index=True)
print(f"\n✓ Total Health Records: {len(df_health):,}")

# Load Air Quality Data
cities = ['New_York', 'Los_Angeles', 'Chicago', 'Houston', 'Phoenix']
air_dfs = []
for city in cities:
    df = pd.read_csv(f'../data/raw/air_quality/{city}_air_quality.csv')
    air_dfs.append(df)

df_air = pd.concat(air_dfs, ignore_index=True)
print(f"✓ Total Air Quality Records: {len(df_air):,}")

# Load Weather Data
weather_dfs = []
for city in cities:
    df = pd.read_csv(f'../data/raw/weather/{city}_weather.csv')
    weather_dfs.append(df)

df_weather = pd.concat(weather_dfs, ignore_index=True)
print(f"✓ Total Weather Records: {len(df_weather):,}")

# ============================================================================
# STEP 2: Debug - Check data formats
# ============================================================================
print("\n" + "=" * 80)
print("[STEP 2] Checking data formats...")
print("=" * 80)

print("\n--- Health Data Sample ---")
print(df_health[['node_id', 'date', 'patient_id']].head())
print(f"\nHealth date type: {df_health['date'].dtype}")
print(f"Unique nodes: {df_health['node_id'].unique()}")

print("\n--- Air Quality Data Sample ---")
print(df_air[['city_id', 'timestamp']].head())
print(f"\nAir timestamp type: {df_air['timestamp'].dtype}")
print(f"Unique cities: {df_air['city_id'].unique()}")

print("\n--- Weather Data Sample ---")
print(df_weather[['city_id', 'timestamp']].head())
print(f"\nWeather timestamp type: {df_weather['timestamp'].dtype}")

# ============================================================================
# STEP 3: Fix date formats and city mappings
# ============================================================================
print("\n" + "=" * 80)
print("[STEP 3] Converting dates and mapping cities...")
print("=" * 80)

# Convert health dates to datetime
df_health['date'] = pd.to_datetime(df_health['date'])
print(f"✓ Converted health dates to datetime")

# Map hospital nodes to city IDs
node_to_city = {
    'hospital_1': 'city_1',
    'hospital_2': 'city_2',
    'hospital_3': 'city_3',
    'hospital_4': 'city_4',
    'hospital_5': 'city_5'
}

df_health['city_id'] = df_health['node_id'].map(node_to_city)
print(f"✓ Mapped hospitals to cities")
print(f"  Unique city_ids in health data: {df_health['city_id'].unique()}")

# Convert air quality timestamps to datetime and extract date
df_air['timestamp'] = pd.to_datetime(df_air['timestamp'])
df_air['date'] = df_air['timestamp'].dt.date
df_air['date'] = pd.to_datetime(df_air['date'])
print(f"✓ Extracted dates from air quality timestamps")

# Convert weather timestamps to datetime and extract date
df_weather['timestamp'] = pd.to_datetime(df_weather['timestamp'])
df_weather['date'] = df_weather['timestamp'].dt.date
df_weather['date'] = pd.to_datetime(df_weather['date'])
print(f"✓ Extracted dates from weather timestamps")

# ============================================================================
# STEP 4: Aggregate data by date and city
# ============================================================================
print("\n" + "=" * 80)
print("[STEP 4] Aggregating data by date and city...")
print("=" * 80)

# Aggregate health data
print("\nAggregating health data...")
health_daily = df_health.groupby(['date', 'city_id']).agg({
    'heart_rate': 'mean',
    'spo2': 'mean',
    'body_temp': 'mean',
    'steps': 'mean',
    'sleep_hours': 'mean',
    'health_status': lambda x: (x == 'sick').sum() / len(x) * 100,  # % sick
    'patient_id': 'count'  # number of patients
}).reset_index()

# Rename columns for clarity
health_daily.rename(columns={
    'health_status': 'sick_percentage',
    'patient_id': 'num_patients'
}, inplace=True)

print(f"✓ Health daily aggregated: {len(health_daily)} records")
print(f"  Date range: {health_daily['date'].min()} to {health_daily['date'].max()}")
print(f"  Cities: {health_daily['city_id'].unique()}")

# Aggregate air quality data
print("\nAggregating air quality data...")
air_daily = df_air.groupby(['date', 'city_id']).agg({
    'pm25': 'mean',
    'pm10': 'mean',
    'no2': 'mean',
    'co': 'mean',
    'o3': 'mean',
    'aqi': 'mean'
}).reset_index()

print(f"✓ Air quality daily aggregated: {len(air_daily)} records")
print(f"  Date range: {air_daily['date'].min()} to {air_daily['date'].max()}")
print(f"  Cities: {air_daily['city_id'].unique()}")

# Aggregate weather data
print("\nAggregating weather data...")
weather_daily = df_weather.groupby(['date', 'city_id']).agg({
    'temperature': 'mean',
    'humidity': 'mean',
    'pressure': 'mean',
    'wind_speed': 'mean',
    'precipitation': 'sum'  # total daily precipitation
}).reset_index()

print(f"✓ Weather daily aggregated: {len(weather_daily)} records")
print(f"  Date range: {weather_daily['date'].min()} to {weather_daily['date'].max()}")
print(f"  Cities: {weather_daily['city_id'].unique()}")

# ============================================================================
# STEP 5: Merge all datasets
# ============================================================================
print("\n" + "=" * 80)
print("[STEP 5] Merging all datasets...")
print("=" * 80)

# Check for overlapping dates and cities before merge
print("\nChecking data compatibility...")
health_keys = set(zip(health_daily['date'], health_daily['city_id']))
air_keys = set(zip(air_daily['date'], air_daily['city_id']))
weather_keys = set(zip(weather_daily['date'], weather_daily['city_id']))

print(f"Health unique (date, city) pairs: {len(health_keys)}")
print(f"Air quality unique (date, city) pairs: {len(air_keys)}")
print(f"Weather unique (date, city) pairs: {len(weather_keys)}")

common_keys = health_keys & air_keys & weather_keys
print(f"Common (date, city) pairs across all datasets: {len(common_keys)}")

# Perform merges
print("\nPerforming inner merge on health + air quality...")
merged = health_daily.merge(air_daily, on=['date', 'city_id'], how='inner')
print(f"✓ After health+air merge: {len(merged)} records")

print("\nPerforming inner merge with weather...")
merged = merged.merge(weather_daily, on=['date', 'city_id'], how='inner')
print(f"✓ After adding weather: {len(merged)} records")

if len(merged) == 0:
    print("\n⚠️ WARNING: Merge resulted in 0 records!")
    print("\nDEBUGGING INFO:")
    
    # Show sample dates from each dataset
    print("\nSample dates from Health:")
    print(health_daily[['date', 'city_id']].head(10))
    
    print("\nSample dates from Air Quality:")
    print(air_daily[['date', 'city_id']].head(10))
    
    print("\nSample dates from Weather:")
    print(weather_daily[['date', 'city_id']].head(10))
    
    # Try left merge to see what's missing
    print("\nTrying LEFT merge to diagnose...")
    test_merge = health_daily.merge(air_daily, on=['date', 'city_id'], how='left', indicator=True)
    print(test_merge['_merge'].value_counts())
    
else:
    print("\n✅ SUCCESS! Merged dataset created")
    
    # ============================================================================
    # STEP 6: Add additional features
    # ============================================================================
    print("\n" + "=" * 80)
    print("[STEP 6] Adding derived features...")
    print("=" * 80)
    
    # Add day of week
    merged['day_of_week'] = merged['date'].dt.dayofweek
    merged['is_weekend'] = merged['day_of_week'].isin([5, 6]).astype(int)
    
    # Add month
    merged['month'] = merged['date'].dt.month
    
    # Add season (simplified for Northern Hemisphere)
    def get_season(month):
        if month in [12, 1, 2]:
            return 'winter'
        elif month in [3, 4, 5]:
            return 'spring'
        elif month in [6, 7, 8]:
            return 'summer'
        else:
            return 'fall'
    
    merged['season'] = merged['month'].apply(get_season)
    
    # Add pollution severity categories
    merged['pollution_level'] = pd.cut(merged['aqi'], 
                                       bins=[0, 50, 100, 150, 200, 300, 500],
                                       labels=['Good', 'Moderate', 'Unhealthy for Sensitive', 
                                              'Unhealthy', 'Very Unhealthy', 'Hazardous'])
    
    # Add temperature categories
    merged['temp_category'] = pd.cut(merged['temperature'],
                                     bins=[-np.inf, 10, 20, 25, 30, np.inf],
                                     labels=['Cold', 'Cool', 'Moderate', 'Warm', 'Hot'])
    
    print("✓ Added derived features:")
    print("  - day_of_week, is_weekend")
    print("  - month, season")
    print("  - pollution_level")
    print("  - temp_category")
    
    # ============================================================================
    # STEP 7: Save merged dataset
    # ============================================================================
    print("\n" + "=" * 80)
    print("[STEP 7] Saving merged dataset...")
    print("=" * 80)
    
    # Reorder columns for better readability
    column_order = [
        'date', 'city_id', 'day_of_week', 'is_weekend', 'month', 'season',
        # Health metrics
        'sick_percentage', 'num_patients', 'heart_rate', 'spo2', 'body_temp', 'steps', 'sleep_hours',
        # Air quality
        'pm25', 'pm10', 'no2', 'co', 'o3', 'aqi', 'pollution_level',
        # Weather
        'temperature', 'temp_category', 'humidity', 'pressure', 'wind_speed', 'precipitation'
    ]
    
    merged = merged[column_order]
    
    # Save to CSV
    output_file = 'data/processed/merged_daily_data.csv'
    merged.to_csv(output_file, index=False)
    print(f"✓ Saved: {output_file}")
    
    # ============================================================================
    # STEP 8: Summary statistics
    # ============================================================================
    print("\n" + "=" * 80)
    print("MERGED DATASET SUMMARY")
    print("=" * 80)
    
    print(f"\n📊 Total Records: {len(merged):,}")
    print(f"📅 Date Range: {merged['date'].min()} to {merged['date'].max()}")
    print(f"🏙️ Cities: {merged['city_id'].nunique()}")
    print(f"📏 Features: {len(merged.columns)}")
    
    print("\n🔢 Dataset Info:")
    print(merged.info())
    
    print("\n📈 Sample Data:")
    print(merged.head(10))
    
    print("\n📊 Basic Statistics:")
    print(merged.describe())
    
    print("\n✅ Merged dataset is ready for modeling!")
    print("=" * 80)
```
